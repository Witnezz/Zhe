```python
import pandas as pd
df = pd.read_csv("companies.csv")
df.head()
df.sort_values(by='annualRevenue',ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.figure();
df.plot(kind="bar");
```

```python
df['location']=df['city']+df['state']+df['country']
print(df)
```

```python
plt.scatter('employees', 'companyName', s=df["employees"]/1000, alpha=0.5,  data=df)
plt.xlabel("employess", size=16)
plt.ylabel("companies", size=16)
plt.title("Bubble Chart by the Number of Employees", size=18)
```

```python
plt.pie(df['annualRevenue'], labels=df['companyName'],
        startangle=90,
        autopct = '%1.2f%%')

plt.axis('equal') 
plt.title('Pie Chart by Revenue')
plt.show()

plt.pie(df['employees'], labels=df['companyName'],
        startangle=90,
        autopct = '%1.2f%%')
plt.axis('equal') 
plt.title('Pie Chart by the Number of Employees')
plt.show()
```
